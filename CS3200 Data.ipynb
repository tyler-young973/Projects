```python
import urllib.request as urllib
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
player_stat = pd.DataFrame(columns = ["Game ID","Name", "Minutes Played", "Points", "Rebounds", "Assists"])
game_stat = pd.DataFrame(columns = ["Home", "Home_Score", "Away_Score", "Date", "Opponent"])
```

```python
player_stat
game_stat
```

```python
html = urllib.urlopen('https://www.basketball-reference.com/teams/BOS/2019_games.html')
soup = BeautifulSoup(html.read())
```

```python
tr_tags = soup.find_all("tr")
links = []
for tag in tr_tags:
    td_tag = tag.find_all("td")
    if td_tag is not None:
        for td in td_tag:
            a_tag = td.find("a")
            if a_tag is not None:
                   if a_tag.get_text() == 'Box Score':
                        links.append('https://www.basketball-reference.com/' + a_tag.get('href'))
```

```python
game_id = 1
for x in links:
    html = urllib.urlopen(x)
    soup = BeautifulSoup(html.read())
    
    
    table_tags = soup.find("div", class_="section_wrapper box-BOS box-BOS-game")
    tr_tag = table_tags.find_all("tr")
    for tag in tr_tag:
        th_tag = tag.find("th")
        if th_tag is not None:
            a_tag = th_tag.find('a')
            if a_tag is not None:
                  player_name = a_tag.get_text()
        td_tag = tag.find('td')
        if td_tag is not None:
            if td_tag.get_text() == 'Did Not Play' or td_tag.get_text() == "Not With Team" or td_tag.get_text() == "Did Not Dress":
                minutes = 0;
                points = 0;
                rebounds = 0;
                assists = 0;

            else:
                td_tag = tag.find_all('td')
                if td_tag is not None:
                    if td_tag[0] is not None:           
                        minutes = td_tag[0].get_text()
                    if td_tag[18] is not None:
                        points = td_tag[18].get_text()
                    if td_tag[12] is not None:
                        rebounds = td_tag[12].get_text()
                    if td_tag[13] is not None:
                        assists = td_tag[13].get_text()
        

        player_stat = player_stat.append({"Game ID": game_id,"Name": player_name, "Minutes Played" : minutes, "Points" : points, "Rebounds": rebounds, 
        "Assists": assists}, ignore_index = True)
    game_id += 1


```
