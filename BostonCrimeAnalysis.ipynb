```python
boston_crime["Severity"] = 0
most_affected_areas = ["Roxbury","Dorchester","South End","Mattapan","Downtown"]
most_affected_times = [17,18,16,12,19]
most_affected_months = [8,7,6,9,5]
most_affected_days = ["Friday","Wednesday","Thursday"]

for index, row in boston_crime.iterrows():
    if(row["UCR_PART"] == 1.0 and row["HOUR"] in most_affected_times and row["MONTH"] in most_affected_months and row["DAY_OF_WEEK"] in most_affected_days):
        boston_crime.loc[index,"Severity"] = 5.0
    elif(row["UCR_PART"] == 1.0):
        boston_crime.loc[index,"Severity"] = 4.0
    elif(row["UCR_PART"] == 2.0 and row["HOUR"] in most_affected_times and row["MONTH"] in most_affected_months and row["DAY_OF_WEEK"] in most_affected_days):
        boston_crime.loc[index,"Severity"] = 3.0
    elif(row["UCR_PART"] == 2.0):
        boston_crime.loc[index,"Severity"] = 2.0
    elif(row["UCR_PART"] == 1.0 and row["HOUR"] in most_affected_times and row["MONTH"] in most_affected_months and row["DAY_OF_WEEK"] in most_affected_days):
        boston_crime.loc[index,"Severity"] = 2.0
    else:
        boston_crime.loc[index,"Severity"] = 1.0
        
```

### Hypothesis Test

```python
def describe_data(data, IV, DV):
    descriptives = data.groupby(IV).agg(["count", "mean", "std", "sem"])
    descriptives = descriptives[DV]
    return descriptives
```

```python
from scipy import stats
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison
```

```python
def oneway_ANOVA(data, IV, DV):
    stats.f_oneway?
    zero = data[data[IV] == 0][DV]
    one = data[data[IV] == 1][DV]
    two = data[data[IV] == 2][DV]
    three = data[data[IV] == 3][DV]
    
    print('-----------------------')
    print('ONE-WAY ANOVA RESULTS')
    print('-----------------------')
      
    print('F-test')
    print('-------')
    
    df1 = len(descriptives) - 1
    df2 = (len(zero) - 1) + (len(one) - 1) + (len(two) - 1) + (len(three) - 1)
    results = stats.f_oneway(zero, one, two, three)
    fstatistic = results[0]
    pvalue = results[1]

    print('F(',df1,',',df2,') =',format(fstatistic, '.2f'), ' p = ', format(pvalue, '.4f'))
  
    print('Assumption Checks')
    print('-------------------')
       
    print('Assumption of Equality of Variances:')
    levene_results = stats.levene(zero, one, two, three)
    print(levene_results)
    
    if levene_results[1] > 0.05:
        print ("Assumption is met. p > .05")
    else :
        print("Assumption is violated. p < .05")
     
    
    print("Assumption of Normality:")
    
    shapiro_dynamic = stats.shapiro(zero)
    print("Dynamic:", shapiro_dynamic)
    
    if shapiro_dynamic[1] > 0.05:
        print ("Assumption is met. p > .05")
    else :
          print("Assumption is violated. p < .05")
    
    
    shapiro_dynamic = stats.shapiro(one)
    print("Dynamic:", shapiro_dynamic)
    
    if shapiro_dynamic[1] > 0.05:
        print ("Assumption is met. p > .05")
    else :
          print("Assumption is violated. p < .05")
    
    shapiro_explicit = stats.shapiro(two)
    print("Explicit:", shapiro_explicit)
    
    if shapiro_explicit[1] > 0.05:
        print ("Assumption is met. p > .05")
    else :
          print("Assumption is violated. p < .05")
    
    shapiro_auto = stats.shapiro(three)
    print("Auto-suggestion:", shapiro_auto)
    
    if shapiro_auto[1] > 0.05:
        print ("Assumption is met. p > .05")
    else :
          print("Assumption is violated. p < .05")
            
    print(' ')       
    print('Post-hoc Tests')
    print('----------------')
    
    mc = MultiComparison(data[DV], data[IV])

    tukey_result = mc.tukeyhsd()
 
    print(tukey_result)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
from sklearn.linear_model import Ridge
estimators = {
    'Linear Regression': LinearRegression(), 
    'kNN': KNeighborsRegressor(),
     'Ridge': Ridge()}
```

```python
def split_the_dataset():
    X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=3000)
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = split_the_dataset()
```

```python
def regressors_percentage_split():
    for estimator_name, estimator_object in estimators.items():
        model = estimator_object.fit(X=X_train, y=y_train)
        print(estimator_name + ":")
        print("\tR-squared value for training set: ", r2_score(y_train, model.predict(X_train)))
        print("\tR-squared value for testing set: ", r2_score(y_test, model.predict(X_test)))
        print("\n")
```

```python
regressors_percentage_split()
```

```python
def preprocessed_regression():
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    scaler.fit(X_train) 
    X_train_scaled = scaler.transform(X_train) 
    X_test_scaled = scaler.transform(X_test) 
    
    for estimator_name, estimator_object in estimators.items():
        model = estimator_object.fit(X=X_train_scaled, y=y_train)
        print(estimator_name + ":")
        print("\tR-squared value for training set: ", r2_score(y_train, model.predict(X_train_scaled)))
        print("\tR-squared value for testing set: ", r2_score(y_test, model.predict(X_test_scaled)))
        print("\n")
        
    return X_train_scaled, X_test_scaled
```

```python
X_train_scaled, X_test_scaled = preprocessed_regression()
```

```python
def SFM_feature_selection():
    from sklearn.feature_selection import SelectFromModel
    from sklearn.tree import DecisionTreeRegressor
    select = SelectFromModel(DecisionTreeRegressor(random_state = 3000), threshold = "median")
    select.fit(X_train, y_train)
    
    X_train_selected = select.transform(X_train_scaled)
    X_test_selected = select.transform(X_test_scaled)
    
    model = KNeighborsRegressor().fit(X=X_train_selected, y=y_train)
    
    print("Selected features after SFM:")
    for feature, arr in zip(features.columns, select.get_support()):
        if arr == True:
            print("\t",feature)
    
    print("kNN Regression performance with selected features:")
    print("\tR-squared value for training set: ", r2_score(y_train, model.predict(X_train_selected)))
    print("\tR-squared value for testing set: ", r2_score(y_test, model.predict(X_test_selected)))
    
    return X_train_selected,X_test_selected   
```

```python
X_train_selected, X_test_selected = SFM_feature_selection()
```

```python
param_grid = {"n_neighbors":[5, 7, 9], "metric": ["euclidean", "manhattan", "minkowski"]}
```

```python
def grid_search_kNN():
    from sklearn.model_selection import GridSearchCV
    grid_search = GridSearchCV(KNeighborsRegressor(), param_grid, cv=5)
    
    grid_search.fit(X=X_train_selected, y=y_train)
    
    print("Best parameters: ", grid_search.best_params_)
    print("Training set score with best parameters: ", grid_search.score(X_train_selected, y_train))
    print("Test set score with best parameters: ", grid_search.score(X_test_selected, y_test))
```

##### Hypothesis Test
* We decided to use a One-way ANOVA test since our categorial variable had 4 levels, and our severity dependent variable is also measured in each condition of each group
* The hypothesis we are trying to solve here is that there is a statistical significance between the effect of a ucr part of a crime towards the severity of a crime
* The null hypothesis would be there are no differences among the ucr part means.
* Results revealed a statistically significant difference among the 3 resaerch methods, F( 3 , 296569 ) =4014627.11, p < 0.001, and this p-value indicates that the null hypothesis can be rejected. Post-hoc comparisons using the Tukey test revealed that the average number of severity of a ucr part of 1 (M = 4.07, SE = .001023) was significantly greater than the ucr parts of 2 (M = 2.07, SE = .000867), 3 (M = 1, SE = 0), and 0 (M = 1, SE = 0). The ucr parts of 3 and 0 did not differ from each other at all compared to the variations between the other ucr parts. 
* Overall these findings indicate that a ucr part of 1 has the highest amount of severity, whereas higher ucr parts would have a lower severity, not including a ucr of 0.
* Ultimately these results did prove did ucr part does have a signficance on determining the severity of the crime, as lower ucr parts (not including a ucr of 0), means a higher level of severity.
* In regards to our data, we would want to have signficant differences since this would mean that our results are not attributed to chance.
* Based on our results, it was determined that these features can in fact predict the severity of the crime, which
